### Basic Graph

```python
import numpy as np
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize the graph by setting the figure size and DPI (dots per inch)
plt.figure(figsize=(5, 3), dpi=100)

# Plot the first line with detailed customization
# The commented line below shows how to set specific parameters:
# - color='red': line color
# - linewidth=2: line thickness
# - marker='*': marker style
# - linestyle='--': dashed line style
# - markersize=10: size of the marker
# - markeredgecolor='blue': color of the marker edge
# plt.plot(x, y, label='2x', color='red', linewidth=2, marker='*', linestyle='--', markersize=10, markeredgecolor='blue')

# Use shorthand notation to plot the same line with a simpler format
# fmt = '[color][marker][line]'
plt.plot(x, y, 'b*--', label='2x')  # 'b*--' means blue color, '*' marker, and dashed line

# Line Number Two: Create a second line segment with different formatting
x2 = np.arange(0, 4.5, 0.5)  # Generate values from 0 to 4.5 in steps of 0.5

# Plot the first part of x^2 (solid red line for values up to index 5)
plt.plot(x2[:6], x2[:6]**2, 'r', label='x^2')

# Plot the second part of x^2 (dashed red line starting from index 5 onward)
plt.plot(x2[5:], x2[5:]**2, 'r--', label='x^2')

# Set the title with font customization
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# Label the x-axis and y-axis
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Define custom ticks on the x-axis and y-axis
plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

# Add a legend to describe the plotted lines
plt.legend()

# Save the figure with high DPI
plt.savefig('ourfirstgraph.png', dpi=220)

# Display the plot
plt.show()
```
